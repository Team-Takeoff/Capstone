```python
import pyarrow.parquet as pq
import pandas as pd
parquet_file = pq.ParquetFile(r'/content/drive/Shareddrives/STUDENT-Capstone SS23/BTS_data/OnTime_Parquet/OnTime_filtered.parquet/part-00000-dc526c93-29d6-4aa1-b7cc-4f555bbee69e-c000.snappy.parquet')
df_comb=None
for i in parquet_file.iter_batches(batch_size=100000):
  df_curr=i.to_pandas()
  if df_comb is None:
    df_comb=df_curr
  else:
    df_comb=pd.concat([df_comb,df_curr])
```

```python
df_comb['FlightDate']=pd.to_datetime(df_comb['FlightDate'])
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df_comb['DepTimeBlk_enc']=encoder.fit_transform(df_comb['DepTimeBlk'])
df_comb['ArrTimeBlk_enc']=encoder.transform(df_comb['ArrTimeBlk'])
```

```python
from sklearn.impute import SimpleImputer
# na_cols=[]
# for col in df_comb.columns:
#   if len(df_comb[df_comb[col].isna()])>0:
#     na_cols.append(col)
remove_cols=['WheelsOff','WheelsOn','FlightDate','DepTimeBlk','ArrTimeBlk']
# na_cols=[col for col in na_cols if col not in remove_cols]
# for col in na_cols:
#   imp=SimpleImputer(strategy='median')
#   df_comb[col]=imp.fit_transform(df_comb[[col]])
#     imp=SimpleImputer(strategy='median')
#     df_comb[col+'_imp']=imp.fit_transform(df_comb[[col]])
imp=SimpleImputer(strategy='median')
df_comb['DepartureDelayGroups']=imp.fit_transform(df_comb[['DepartureDelayGroups']])
```

```python
df_comb['Year']=df_comb['FlightDate'].dt.year
df_comb['Month']=df_comb['FlightDate'].dt.month
df_comb['Day']=df_comb['FlightDate'].dt.day
```
