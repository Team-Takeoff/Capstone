## Join all annual weather files into one for each year
<br>
Currently there's a weather file for each weather station matching an airport for each year of data. Almost all airports - several didn't have a match and will not have weather data associated.  The Airport Code was previously added as a column to each file.<br>
<br>
This code reads each years set of files that match to airport codes in the OnTime data and merges them together to form yearly weather files. These pickled files end up being about 5 GB each.
