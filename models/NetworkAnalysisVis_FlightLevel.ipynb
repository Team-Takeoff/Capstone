```
# Extract the Infected and Recovered counts for each timestep
def preprocess_delay_type_list(delay_list):
  DelayTypeList=[]
  for i, delay in enumerate(delay_list):
    if isinstance(delay, str):
      if delay in ['W', 'P', 'WP']:
        DelayTypeList.append('I')
      elif delay == 'na':
        if (i > 0 and delay_list[i - 1] in ['W', 'P', 'WP']) and delay=='na':
          DelayTypeList.append('R')
        else:
          DelayTypeList.append('S')  # 'na' before an infection is considered as no infection

  return DelayTypeList

df_sorted['SIR_Counts']=df_sorted.DelayTypeList.apply(preprocess_delay_type_list)
df_sorted=df_sorted[df_sorted.Flights>1]
#  [['Adj_Date_Str','OrigList','DestList','DelayTypeList','SIR_Counts']]
```

```
df_sorted['Adj_Date_Str']=pd.to_datetime(df_sorted.Adj_Date_Str)
df_sorted.set_index('Adj_Date_Str', inplace=True)
df_sorted.reset_index(inplace=True)
df_sorted_exp=df_sorted[['Adj_Date_Str','OrigList','DestList','DelayTypeList','SIR_Counts']]
df_exploded=df_sorted_exp.apply(pd.Series.explode)
```

```
df_exploded['month']=df_exploded.Adj_Date_Str.dt.month
df_exploded['year']=df_exploded.Adj_Date_Str.dt.year
```

```
# adding a hub indicator
df_hubs=pd.read_csv('./BTS_data/Hubs.csv')
hubs=df_hubs.Code.to_list()
df_exploded.loc[df_exploded.OrigList.isin(hubs), ['OrigHub']] = True
df_exploded.loc[df_exploded.DestList.isin(hubs), ['DestHub']] = True
df_exploded.OrigHub.fillna(False,inplace=True)
df_exploded.DestHub.fillna(False,inplace=True)
# df_exploded['Dest_hubs']=df_exploded.loc[df_exploded.DestList.isin(hubs)]
df_exploded
```

# Visualize the SIR rates - pyvis

```
G = nx.DiGraph()
# loop through the df rows and add nodes and edges to the graph
for idx, row in df_exploded.iterrows():
  # adding origin node with attributes DelayType and SIR_Counts as dictionaries
  G.add_node(row['OrigList'], DelayType={}, SIR_Counts={},lat=row['Orig_Lat'], lon=row['Orig_Lon'], clustering_coef=None, bottleneck=False)
  # adding destination node with attributes DelayType and SIR_Counts as dictionaries
  G.add_node(row['DestList'], DelayType={}, SIR_Counts={},lat=row['Dest_Lat'], lon=row['Dest_Lon'], clustering_coef=None, bottleneck=False)
  # adding edge from origin to destination
  G.add_edge(row['OrigList'], row['DestList'],weight=row['Weight'])

# count up the delay type and SIRs
for idx, row in df_exploded.iterrows():
  # count the DelayType for the origin node
  delay_type_origin = row['DelayTypeList']
  G.nodes[row['OrigList']]['DelayType'][delay_type_origin] = G.nodes[row['OrigList']]['DelayType'].get(delay_type_origin, 0) + 1

  # count the SIR_Counts for the origin node
  sir_counts_origin = row['SIR_Counts']
  G.nodes[row['OrigList']]['SIR_Counts'][sir_counts_origin] = G.nodes[row['OrigList']]['SIR_Counts'].get(sir_counts_origin, 0) + 1

  # count the DelayType for the destination node
  delay_type_dest = row['DelayTypeList']
  G.nodes[row['DestList']]['DelayType'][delay_type_dest] = G.nodes[row['DestList']]['DelayType'].get(delay_type_dest, 0) + 1

  # count the SIR_Counts for the destination node
  sir_counts_dest = row['SIR_Counts']
  G.nodes[row['DestList']]['SIR_Counts'][sir_counts_dest] = G.nodes[row['DestList']]['SIR_Counts'].get(sir_counts_dest, 0) + 1
```

### pagerank as a mean to identify delay bottleneck

```
def pagerank_with_delay(G, alpha=0.85, iterations=100, delay_factor=0.1):
  G_delayed = G.copy()
  # Assign delays to edges based on the delay_factor and DelayType
  # alpha controls the balance between the influence of the airport's direct connections (edges) and the random jumps to other airports. standard alpha range from 0.8 to 0.9.
  # delay factor controls the impact of delays on the edge weights, increasing this can simulate delays being propagated through the network and influence the ranking
  # of airports based on their susceptibility to delays.
  for u, v, data in G_delayed.edges(data=True):
    if 'DelayType' in data and data['DelayType']:
      delay_factor_edge = delay_factor * len(data['DelayType'])
      data['weight'] = 1 / (data['weight'] + delay_factor_edge)
    else:
      data['weight'] = 1 / data['weight']
  # pagerank with delays
  pagerank_delayed = nx.pagerank(G_delayed, alpha=alpha, max_iter=iterations, weight='weight')

  return pagerank_delayed
```

A higher delay factor will make delays have a more substantial impact on the rankings, while adjusting alpha will control the balance between local connectivity and random jumps when calculating the influence of each airport.

```
def bottleneck_analysis(G, pagerank_scores_before_delay, pagerank_scores_after_delay):
  bottleneck_airports = []
  # difference in PageRank scores
  for airport in G.nodes():
    pr_before_delay = pagerank_scores_before_delay[airport]
    pr_after_delay = pagerank_scores_after_delay[airport]
    pr_difference = pr_after_delay - pr_before_delay

    # find airports with significant increase in delay spread (bottlenecks)
    if pr_difference > 0.001:  # threshold can be changed
      bottleneck_airports.append((airport, pr_difference))

  # sort the airports based on the increase in delay spread
  bottleneck_airports.sort(key=lambda x: x[1], reverse=True)
  return bottleneck_airports

# PageRank scores before and after delay propagation
pagerank_scores_before_delay = nx.pagerank(G, alpha=0.85)
pagerank_scores_after_delay = pagerank_with_delay(G)

# bottleneck analysis
bottleneck_airports = bottleneck_analysis(G, pagerank_scores_before_delay, pagerank_scores_after_delay)

# Print the top bottleneck airports and their increase in delay spread
print("Top Bottleneck Airports:")
print(bottleneck_airports[:10])
delay_propagate_pagerank=[airport for airport, score in bottleneck_airports[:10]]
```

```
import networkx as nx
import numpy as np
import random
import plotly.graph_objects as go

# Set simulation parameters
simulation_steps = 10
beta = 0.0767468104597565  # average
gamma = 0.9156015085033505  # Recovery rate


# Randomly select 20 nodes to be initially infected
initial_infected_nodes = random.sample(list(G.nodes), 100)

# Initialize S, I, and R counts for all nodes
SIR_counts = np.zeros((len(G), 3))  # 3 columns for S, I, and R
SIR_counts[:, 0] = 1  # Set initial susceptible count for all nodes (1 for each node)

# Convert node names to their corresponding integer indices in SIR_counts array
node_index_map = {node: idx for idx, node in enumerate(G.nodes)}
initial_infected_indices = [node_index_map[node] for node in initial_infected_nodes]
SIR_counts[initial_infected_indices, 1] = 1  # Set initial infected count for selected nodes

# Simulate the SIRS model
susceptible_counts = [SIR_counts[:, 0].sum()]
infectious_counts = [SIR_counts[:, 1].sum()]
recovered_counts = [SIR_counts[:, 2].sum()]

for _ in range(simulation_steps):
    # Get the neighbors of each node and convert them to integer indices
    num_infected_neighbors = np.zeros(len(G))
    for node in G.nodes:
        neighbors = list(G.neighbors(node))
        neighbor_indices = [node_index_map[n] for n in neighbors]
        num_infected_neighbors[node_index_map[node]] = np.sum(SIR_counts[neighbor_indices, 1])

    # Calculate the rate of change using SIRS equations
    dSdt = -beta * SIR_counts[:, 0] * num_infected_neighbors / len(G) + gamma * SIR_counts[:, 2]
    dIdt = beta * SIR_counts[:, 0] * num_infected_neighbors / len(G) - gamma * SIR_counts[:, 1]
    dRdt = gamma * SIR_counts[:, 1]

    # Update the counts for the next time step
    SIR_counts[:, 0] += dSdt
    SIR_counts[:, 1] += dIdt
    SIR_counts[:, 2] += dRdt

    # adding  the results to the lists
    susceptible_counts.append(SIR_counts[:, 0].sum())
    infectious_counts.append(SIR_counts[:, 1].sum())
    recovered_counts.append(SIR_counts[:, 2].sum())

fig = go.Figure()

# Add traces for S, I, and R
fig.add_trace(go.Scatter(x=list(range(simulation_steps + 1)), y=susceptible_counts, mode='lines', name='Susceptible'))
fig.add_trace(go.Scatter(x=list(range(simulation_steps + 1)), y=infectious_counts, mode='lines', name='Infectious'))
fig.add_trace(go.Scatter(x=list(range(simulation_steps + 1)), y=recovered_counts, mode='lines', name='Recovered'))

# Update the layout
fig.update_layout(title='SIRS Model Simulation',
                  xaxis_title='Time Step',
                  yaxis_title='Number of Individuals',
                  legend=dict(x=0, y=1),
                  )

# Show the plot
fig.show()
```
